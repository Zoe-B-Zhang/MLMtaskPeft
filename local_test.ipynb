```python
import torch

def get_gpu_memory_size():
    if torch.cuda.is_available():
        device = torch.device("cuda")
        memory_size = torch.cuda.get_device_properties(device).total_memory
        memory_size_gb = memory_size / (1024 ** 3)  # Convert bytes to gigabytes
        return memory_size_gb
    else:
        return 0

# Get the GPU memory size
gpu_memory_size = get_gpu_memory_size()
print(f"GPU memory size: {gpu_memory_size:.2f} GB")
```
